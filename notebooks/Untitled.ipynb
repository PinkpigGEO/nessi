# SU/CWP format support

NeSSI uses the SU/CWP format as internal structure for manipulating seismic data. 

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Import SU/CWP format support
from nessi.io import SUdata
```

```python
# Initialize SU/CWP object
dobs = SUdata()

# Read the SU/CWP file
dobs.read('data/musc_F50_01.su')

# Plot SU traces
dobs.image(clip=1.0, legend=1, label1='Time [m/s]', label2='Receiver ID', title='Original data')
plt.show()
```

```python
# Windowing

dobs.wind(tmin=0., tmax=0.3)

# Plot the SU traces
dobs.image(clip=1.0, legend=1, label1='Time [m/s]', label2='Receiver ID', title='Windowed data')
plt.show()
```
